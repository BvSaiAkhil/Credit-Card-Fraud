```python
x=data_set.iloc[:,1:30].values
y=data_set.iloc[:,30].values
```

```python
from sklearn.cross_validation import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
#Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
LRmodel=LogisticRegression(random_state=0)
```

```python
LRmodel.fit(x_train,y_train)
```

```python
y_predLR=LRmodel.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm_LR=confusion_matrix(y_test,y_predLR)
```

```python
plt.figure(figsize=(15,7))
data_frameLR=pd.DataFrame(cm_LR,index=['Not Fraud','Fraud'],columns=['Not Fraud','Fraud'])
sns.heatmap(data_frameLR,cmap='Blues',annot=True,fmt='g')
plt.show()
```

```python
from sklearn.metrics import f1_score,recall_score
```

```python
f1_scoreLR=f1_score(y_test,y_predLR)
recall_scoreLR=recall_score(y_test,y_predLR)
print('F1 score for Logistic Regression Model:{}'.format(f1_scoreLR))
print('Recall score for Logistic Regression Model:{}'.format(recall_scoreLR))
```

```python
#KNN Model
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
y_predknn=knn.predict(x_test)
```

```python
cm_knn=confusion_matrix(y_test,y_predknn)
```

```python
plt.figure(figsize=(15,7))
data_frameknn=pd.DataFrame(cm_knn,index=['Not Fraud','Fraud'],columns=['Not Fraud','Fraud'])
sns.heatmap(data_frameknn,cmap='Blues',annot=True,fmt='g')
plt.show()
```
